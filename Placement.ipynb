```python
new_data['hsc_b'].unique()
```

```python
new_data['hsc_b'] = new_data['hsc_b'].map({'Central':1, 'Others':0})
```

```python
new_data['hsc_s'].unique()
```

```python
new_data['hsc_s'] = new_data['hsc_s'].map({'Science':2, 'Commerce':1, 'Arts':0})
```

```python
new_data.head()
```

```python
new_data['degree_t'].unique()
```

```python
new_data['degree_t'] = new_data['degree_t'].map({'Sci&Tech':2, 'Comm&Mgmt':1, 'Others':0})
```

```python
new_data.head()
```

```python
new_data['specialisation'].unique()
```

```python
new_data['specialisation'] = new_data['specialisation'].map({'Mkt&HR':1, 'Mkt&Fin':0})
```

```python
new_data.head()
```

```python
new_data['workex'].unique()
```

```python
new_data['workex'] = new_data['workex'].map({'No':1, 'Yes':0})
```

```python
new_data.head()
```

```python
new_data['status'].unique()
```

```python
new_data['status'] = new_data['status'].map({'Placed':1, 'Not Placed':0})
```

```python
new_data.head()
```

```python
X = new_data.drop('status', axis = 1)
Y = new_data['status']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 49)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)

sm = SVC()
sm.fit(X_train, Y_train)

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)

rf = RandomForestClassifier()
rf.fit(X_train, Y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train, Y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = sm.predict(X_test)
y_pred3 = knn.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rf.predict(X_test)
y_pred6 = gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score1 = accuracy_score(Y_test, y_pred1)
score2 = accuracy_score(Y_test, y_pred2)
score3 = accuracy_score(Y_test, y_pred3)
score4 = accuracy_score(Y_test, y_pred4)
score5 = accuracy_score(Y_test, y_pred5)
score6 = accuracy_score(Y_test, y_pred6)
```

```python
print('Accuracy of Logistic Regression is :',score1)
print('Accuracy of SVM is :',score2)
print('Accuracy of KNN is :',score3)
print('Accuracy of Decision Tree is :',score4)
print('Accuracy of Random Forest is :',score5)
print('Accuracy of Gradient Boosting is :',score6)
```

```python
final = pd.DataFrame({'Models':['LR', 'SVM', 'KNN', 'DT', 'RF', 'GB'],
        'Accuracy':[score1*100,
                     score2*100,
                     score3*100,
                     score4*100,
                     score5*100,
                     score6*100]})
```
